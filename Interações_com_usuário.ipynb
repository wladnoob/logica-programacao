## Type

* Para vermos a tipagem da variável, se usa o comando type

```python
x = 10
type (x)
print ("x", type(x))

nome = "Wladimir"
print (nome, type (nome))

num = "100"
num_convert = int(num) # Converter variável
print ("num", type(num))
print ("num_convert", type(num_convert))

num_float = float(num)
print ("num_float =", num_float)
```

## Input

* Armazena a informação digitada pelo usuário em uma variável
* O dado armazenado sempre será um texto

```python
nome = input("Digite o nome do usuário: ")
print (nome)
```

```python
valor = input("Digite a nota: ")
print (nome, ", sua nota foi:", valor)
```

```python
valor = input("Digite o valor do produto: ")
print ("Valor:", valor)
```

```python
desconto = valor - 5

# O erro ocorre porque a variável valor é do tipo string, e está tentando subtrair um número inteiro diretamente dela.
# Para corrigir isso, você precisa converter valor para um número inteiro antes de realizar a subtração.
```

```python
valor_convert = int(valor)
desconto = valor_convert - 5
print (desconto)

# Ou

desconto = int(valor) - 5
print (desconto)
```
