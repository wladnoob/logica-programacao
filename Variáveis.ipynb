# Variáveis

*   O uso da igualdade simples (=) indica que será feita uma atribução
*   Se trata de uma situação "case sensitive", maiúsculo e minúsculo importam
*   Começar com letras ou sublinhados
*   Usar nomes que definam o que ela faz no programa (soma ou soma_total)
*   Não usar nomes longos
*   O Python define o tipo de dados na declaração
*   Existem 3 tipos de variáveis: int (número inteiro), float (número real com ponto) e string (texto)

### Exemplos de recomendações

```python
# Exemplo básico: nome da variável é "x" que vai armazenar o valor 10 em um espaço de memória

x = 10
print (x)
```

```python
# Exemplos de nomeações

soma = 2 + 5
print (soma)

soma_total = 2 + 5
print (soma_total)
```

```python
num_int = 10
num_float = 10.5
string_nome = "Wladimir"

print (num_int)
print (num_float)
print (string_nome)
```
